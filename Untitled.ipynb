```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('car_dataset.csv')
df.head()
```

```python
df['Car_Name'].unique()
```

```python
df.shape
```

```python
df['Fuel_Type'].unique()
```

```python
dummies=pd.get_dummies(df['Fuel_Type'],drop_first=True).astype(int)
```

```python
df=pd.concat([df,dummies],axis='columns')
```

```python
df.head()
```

```python
df['Seller_Type'].unique()
```

```python
df['Seller_Type']=df['Seller_Type'].apply(lambda x:0 if x=='Dealer' else 1)
```

```python
df.head()
```

```python
df['Transmission'].unique()
```

```python
df['Transmission']=df['Transmission'].apply(lambda x: 0 if x=='Manual' else 1)
```

```python
df.head()
```

```python
df=df.drop(['Car_Name','Fuel_Type'],axis='columns')
```

```python
df.head()
```

```python
df['Selling_Price'].describe()
```

```python
df['Present_Price'].describe()
```

```python
m=df['Selling_Price'].mean()
st=df['Selling_Price'].std()
```

```python
df=df[(df['Selling_Price'] <(m + 2*st)) | (df['Selling_Price']> (m - 2*st))]
```

```python
df.isna().sum()
```

```python
df=df[(df['Present_Price'] >= (m - 2*st)) & (df['Present_Price'] <= (m + 2*st))]
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
X=df.drop(['Selling_Price'],axis='columns')
y=df['Selling_Price']
```

```python
X.head()
```

```python
 X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
